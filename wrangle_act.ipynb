In the cells below,  three pieces of data 'twitter_archive_enhanced.csv', 'image_predictions.tsv', and 'tweet_json.txt' were gathered for this project and loaded into the notebook. The methods required to gather each data were different.
1. Directly downloaded the WeRateDogs Twitter archive data (twitter_archive_enhanced.csv)

```python
# Importing essential packages.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_TwitterArchive = pd.read_csv('twitter_archive_enhanced.csv') #Reading the twitter_archive_enhanced.csv file.
```

2. Used the Requests library to download the tweet image prediction (image_predictions.tsv)

```python
# Importing essential packages.
import requests
from PIL import Image
import os

folder_name = 'DogsTweetImages'
filename = 'image_predictions.tsv'
# Make directory if it doesn't already exist
if not os.path.exists(folder_name):
    os.makedirs(folder_name)

url = 'https://d17h27t6h515a5.cloudfront.net/topher/2017/August/599fd2ad_image-predictions/image-predictions.tsv'

r = requests.get(url)
with open(folder_name + '/' + filename, 'wb') as f:
        f.write(r.content)
        
df_ImagePredict = pd.read_csv('DogsTweetImages/image_predictions.tsv', sep = '\t')       
df_ImagePredict.head(10)
```

3. Use the Tweepy library to query additional data via the Twitter API (tweet_json.txt)

```python
# Importing essential package.
import json

# Importing the tweet_json.txt from the udacity archive and selecting necessary columns.
df_TweetApi = pd.read_json("tweet_json.txt", lines=True)
df_TweetApi = df_TweetApi[['id', 'retweet_count', 'favorite_count']]

df_TweetApi = df_TweetApi.rename(columns = {'id' : 'tweet_id'})

df_TweetApi.head(10)
```

## Cleaning Data
In this section, clean **all** of the issues you documented while assessing.

### Quality Issue #1: 

#### Define: Incomplete columns with NaN inputs, examples are 'in_reply_to_status_id', 'in_reply_to_user_id', 'retweeted_status_id' and so on should be dropped.

#### Code

```python
# Dropping columns that have greater number of NaN values and that are unnecessary to the study.
df_TwitterArchive_clean.drop(['in_reply_to_status_id', 'in_reply_to_user_id', 'retweeted_status_id',  
                              'retweeted_status_user_id', 'retweeted_status_timestamp', 'expanded_urls'], axis=1, inplace=True)
```

#### Test

### Quality Issue #4:

#### Define: P1, P2, and P3 columns have rows with names starting in lowercase for some and uppercase for others.

### Quality Issue #5:

```python
# Replacing all the _ with spaces in words in p1, p2, and p3 columns using the general column.str.replace() function.

df_ImagePredict_clean.p1 = df_ImagePredict_clean.p1.str.replace('_',' ')
df_ImagePredict_clean.p2 = df_ImagePredict_clean.p2.str.replace('_',' ')
df_ImagePredict_clean.p3 = df_ImagePredict_clean.p3.str.replace('_',' ')
```

### Tidiness Issue #1:

#### Define: Separate doggo, floofer, pupper and puppo columns which should be combined as one column in df_TwitterArchive.

#### Code

```python
# Using the str.extract method tp merge the 'doggo','floofer','pupper','puppo', columns in one column: 'dogtypes'

df_TwitterArchive_clean['dogtypes'] =  df_TwitterArchive_clean['text'].str.extract('(doggo|floofer|pupper|puppo)')
df_TwitterArchive_clean.drop(['doggo','floofer','pupper','puppo'], axis=1, inplace=True)
```
